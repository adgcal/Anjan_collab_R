# PCA based on sequence 


```r
## Function 1 : A function that aggregates residue properties like charge hydrophobicity etc.
```

```r
pro<-function(seq){

  c<-charge(seq, pH = 7, pKscale = "EMBOSS")
  p<-boman(seq)
  a<-hmoment(seq, angle = 100, window = 11)
  h<-hydrophobicity(seq, scale = "KyteDoolittle")
  p<-pI(seq,pKscale = "EMBOSS")
  mw<-mw(seq, monoisotopic = FALSE)
  pro<-cbind(c,p,a,h,p,mw)
  colnames(pro)<-c("Chrg","P-P","Ampi","Hydro","Isoelec","Mwt")
  return(pro)}

```

## A function that evaulates Sequence Based on PCA

```r
npro.pca<-function(seq){

  c<-charge(seq, pH = 7, pKscale = "EMBOSS")
  p<-boman(seq)
  a<-hmoment(seq, angle = 100, window = 11)
  h<-hydrophobicity(seq, scale = "KyteDoolittle")
  p<-pI(seq,pKscale = "EMBOSS")
  mw<-mw(seq, monoisotopic = FALSE)
  pro<-cbind(c,p,a,h,p,mw)
  colnames(pro)<-c("Chrg","P-P","Ampi","Hydro","Isoelec","Mwt")
  prc.pro<-PCA(pro,graph=FALSE)
  return(prc.pro)}
```

# Notes on PCA



PCA (Principal Component Analysis) is a commonly used technique in data analysis and data visualization. It is used to identify patterns in high-dimensional datasets by reducing the number of variables in the dataset while retaining the maximum amount of information possible.

During a PCA analysis, several components are generated. These components are sorted in descending order based on the amount of variance they explain in the dataset. The first component explains the maximum amount of variance, followed by the second component and so on.

In PCA, the "cos2" value (square of the cosine) represents the quality of representation of each variable (or attribute) in the PC (Principal Component). It is a measure of how well the variable is represented in the component. A high value of cos2 means that the variable is well-represented in the component and contributes significantly to the component's variability.

The "screeplot" is a graph that shows the amount of variance explained by each component in descending order. It is used to determine the number of components that should be retained in the analysis. The screeplot displays the eigenvalues of each component against its number. The eigenvalue represents the amount of variance explained by each component. The components with high eigenvalues contribute more to the overall variance of the dataset, and hence, they are considered more important.

In summary, cos2 measures the quality of representation of each variable in the principal component, while the screeplot helps in determining the number of principal components to be retained in the analysis based on the amount of variance they explain.
