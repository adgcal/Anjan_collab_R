## Who first used the IRIS data set 
The iris dataset was introduced by British statistician and biologist Sir Ronald A. Fisher in his 1936 paper "The use of multiple measurements in taxonomic problems" which was published in the journal Annals of Eugenics. Fisher used the iris dataset to demonstrate the effectiveness of linear discriminant analysis for classification tasks. Since then, the iris dataset has become a popular benchmark dataset for testing and comparing classification algorithms in machine learning

## IRIS details 
The iris dataset is one of the most famous datasets in machine learning and is often used as a benchmark dataset for classification algorithms. It contains measurements of the sepal length, sepal width, petal length, and petal width for 150 iris flowers, with 50 samples from each of three iris species (setosa, versicolor, and virginica).

The dataset is included in the base R distribution, so no additional packages are needed to access it. It is often used to demonstrate the power of machine learning algorithms, as it is a relatively simple dataset that can be easily visualized and understood.

The iris dataset is a great resource for beginners in machine learning, as it is easy to work with and is well-documented. However, it is important to note that it is a relatively small dataset and may not represent the complexities of real-world problems.

# Structured form of the same code: USE OF FUNCTION 

```r
# Define a function to filter data based on sex and age
filter_data <- function(data, sex, age) {
  data %>% filter(Sex == sex, Age == age)
}

# Define a function to calculate the survival rate
calc_survival_rate <- function(lived_data, total_data) {
  sum(lived_data$Freq) / sum(total_data$Freq) * 100
}

# Filter the data based on sex and age
female_Child <- filter_data(Titanic, "Female", "Child")
male_Child <- filter_data(Titanic, "Male", "Child")
female <- filter_data(Titanic, "Female", "Adult")
male <- filter_data(Titanic, "Male", "Adult")

# Filter the data based on sex, age, and survival
female_Child_Lived <- filter_data(Titanic, "Female", "Child") %>% filter(Survived == 'Yes')
male_Child_Lived <- filter_data(Titanic, "Male", "Child") %>% filter(Survived == 'Yes')
female_sur <- filter_data(Titanic, "Female", "Adult") %>% filter(Survived == 'Yes')
male_sur <- filter_data(Titanic, "Male", "Adult") %>% filter(Survived == 'Yes')

# Calculate the survival rates
female_Child_Survival_Rate <- calc_survival_rate(female_Child_Lived, female_Child)
male_Child_Survival_Rate <- calc_survival_rate(male_Child_Lived, male_Child)
male_Survival_Rate <- calc_survival_rate(male_sur, male)
female_Survival_Rate <- calc_survival_rate(female_sur, female)

# Create data frame
df <- data.frame(
  Category = c("Female Children", "Male Children", "Female Adults", "Male Adults"),
  Survived_percentage = c(round(female_Child_Survival_Rate, 2), round(male_Child_Survival_Rate, 2),
               round(female_Survival_Rate, 2), round(male_Survival_Rate, 2)),
  Total = c(sum(female_Child$Freq), sum(male_Child$Freq), sum(female$Freq), sum(male$Freq)),
  Survived_Total = c(sum(female_Child_Lived$Freq), sum(male_Child_Lived$Freq),
                     sum(female_sur$Freq), sum(male_sur$Freq))
)

# Display table
knitr::kable(df, align = "c")
```

# Another Structured form of the TITANIC data : Use of *lapply()*

```r
library(dplyr)

# Subset the data into different groups
groups <- list(
  female_Child = Titanic %>% filter(Sex == 'Female' & Age == 'Child'),
  male_Child = Titanic %>% filter(Sex == 'Male' & Age == 'Child'),
  female_Adult = Titanic %>% filter(Sex == 'Female' & Age == 'Adult'),
  male_Adult = Titanic %>% filter(Sex == 'Male' & Age == 'Adult')
)

# Use lapply to calculate survival rate for each group
survival_rates <- lapply(groups, function(x) {
  lived <- x %>% filter(Survived == 'Yes')
  total_freq <- sum(x$Freq)
  lived_freq <- sum(lived$Freq)
  rate <- round(lived_freq / total_freq * 100, 2)
  return(rate)
})

# Combine results into a data frame
df <- data.frame(
  Category = names(survival_rates),
  Survived_percentage = unlist(survival_rates),
  Total = sapply(groups, function(x) summarise(x, total = sum(Freq))$total),
  Survived_Total = sapply(groups, function(x) summarise(x, survived = sum(Freq[Survived == 'Yes']))$survived)
)

# Display table
knitr::kable(df, align = "c")
```

The **ANOVA** table shows the results of an analysis of variance (ANOVA) for the InsectSprays dataset. 
- In this analysis, the response variable is the number of insects counted
- Explanatory variable is the type of insecticide spray used.

- The ANOVA table has two main parts: the "Between-groups" section, which shows the results of the analysis for the explanatory variable (spray), and the "Within-groups" section, which shows the results for the residual error (or unexplained variation).

- In the "Between-groups" section, the table shows the following information:

  - Df: degrees of freedom, which represent the number of groups minus 1 (in this case, 5-1 = 4).
  - Sum Sq: sum of squares, which represent the variation explained by the groups (or spray types).
  - Mean Sq: mean sum of squares, which is the sum of squares divided by the degrees of freedom.
  - F value: the F statistic, which is the ratio of the mean sum of squares for the groups to the mean sum of squares for the residual error. It measures whether the variation between groups is larger than the variation within groups.
- Pr(>F): the p-value, which is the probability of observing an F value as extreme or more extreme than the one calculated, assuming that the null hypothesis (that there is no difference between the means of the groups) is true. A p-value less than 0.05 is typically considered statistically significant.

In this case, the F value is 34.7 and the p-value is less than 2e-16 (a very small number), which indicates that *there is a significant difference in the effectiveness of the different insecticides* in controlling the number of insects. The "***" symbols next to the p-value indicate that the result is highly significant.

In the "Within-groups" section, the table shows the following information:

- Df: degrees of freedom, which represent the total sample size minus the number of groups (in this case, 72-5 = 67).
- Sum Sq: sum of squares, which represent the variation that is not explained by the groups (or spray types).
- Mean Sq: mean sum of squares, which is the sum of squares divided by the degrees of freedom.
- The residuals represent the variation that is not explained by the model and can be seen as the unexplained variation.

# Interpretation of the lm() based machine learning model:
This output is the result of fitting a linear regression model to the *trainData dataset* using the lm function in R. Here is how to interpret the output:

- The first line of the output (lm(formula = count ~ ., data = trainData)
) indicates the model formula that was used to fit the model. 

- The "Residuals" section of the output provides some summary statistics about the residuals (the differences between the observed values of count and the predicted values from the model). The minimum residual is -9.333, the first quartile (25th percentile) is -2.013, the median (50th percentile) is -0.500, the third quartile (75th percentile) is 2.100, and the maximum residual is 8.222.


- The "Coefficients" section of the output provides estimates of the regression coefficients for each predictor variable in the model, as well as their standard errors, t-values, and p-values. 
- The intercept term (corresponding to the mean count when all other predictors are equal to zero) is estimated to be 14.7778 with a standard error of 1.3324.
-  Each of the spray variables (sprayB, sprayC, sprayD, sprayE, and sprayF) has an estimated coefficient that represents the change in count associated with a one-unit increase in the corresponding variable, holding all other variables constant. 
- For example, the estimated coefficient for sprayC is -12.6528, which means that on average, counts are 12.6528 units lower when sprayC is used compared to when no spray is used, holding all other variables constant.
-  The t-values and p-values for each coefficient indicate whether the coefficient is statistically significant. 
- In this case, the intercept and the sprayC, sprayD, and sprayE coefficients are highly significant (p < 0.001), while the sprayB and sprayF coefficients are not significant at the 0.05 level.

- The "Residual standard error" is an estimate of the standard deviation of the residuals, and provides a measure of how well the model fits the data. In this case, the residual standard error is 3.997, which means that the typical difference between observed and predicted counts is around 4.

- The "Multiple R-squared" and "Adjusted R-squared" values indicate the proportion of variance in the response variable (count) that is explained by the model. In this case, the multiple R-squared is 0.7297, which means that the model explains around 73% of the variance in count. The adjusted R-squared takes into account the number of predictor variables in the model, and is slightly lower at 0.6997.

- The "F-statistic" and associated p-value test whether any of the predictor variables in the model are jointly significant in explaining the response variable. In this case, the F-statistic is 24.3 with 5 and 45 degrees of freedom, and the p-value is very small (p < 0.001), indicating that the model as a whole is highly significant.
